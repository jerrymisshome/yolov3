```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
#img = cv2.imread('/content/perspective_school_2.jpg')
img = cv2.cvtColor(cv2.imread("/content/perspective_school_2.jpg"), cv2.COLOR_BGR2RGB)

rows, cols = img.shape[:2]
```

```python
# 定义平移矩阵，需要是numpy的float32类型
# x轴平移100，y轴平移50
M = np.float32([[1, 0, 100], [0, 1, 50]])
# 用仿射变换实现平移，第三个参数为dst的大小
shift_dst = cv2.warpAffine(img, M, (cols, rows))
plt.figure(figsize=(20,20))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(shift_dst),plt.title('Output')
plt.show()
```

```python
M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
rotate_dst = cv2.warpAffine(img,M,(cols,rows))
plt.figure(figsize=(20,20))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(rotate_dst),plt.title('Output')
plt.show()
```

```python
flip_dst = cv2.flip(img,1)

plt.figure(figsize=(20,20))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(flip_dst),plt.title('Output')
```

```python


pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])

M = cv2.getAffineTransform(pts1,pts2)

affine_dst = cv2.warpAffine(img,M,(cols,rows))
plt.figure(figsize=(20,20))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(affine_dst),plt.title('Output')
plt.show()
```
